```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from smtools import sm_sort
```

```python
x = np.linspace(0, 100, 21)
y1 = x + 0.0001*x**3+15
y2 = x[::-1] + 2
y3 = x + 40.5
# Stark map eigenvalues are usually sorted, obscuring exact crossings. 
dat = np.array([y1, y2, y3]).T
dat = np.sort(dat, axis=1)

fig, ax = plt.subplots()
for ix in range(3):
    ax.plot(x, dat[:,ix], marker='.')
plt.show()
```

```python
dat2 = sm_sort(dat)

fig, ax = plt.subplots()
for ix in range(3):
    ax.plot(x, dat2[:,ix], marker='.')
plt.show()
```
